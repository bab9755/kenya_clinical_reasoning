```python
import re
import string

def clean_clinician_text(text): #we are going to clean the clinician response
    if not isinstance(text, str):
        return ""

    text = text.lower()
    text = text.replace('\n', " ")
    text = text.replace('\r', " ")
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r'\s+', " ", text).strip()

    return text

train_df['Cleaned_Clinician'] = train_df['Clinician'].apply(clean_clinician_text) # already clean but just doing it for practice


print("\nOriginal Clinician response (first entry):")
print(train_df['Clinician'].iloc[0])
print("\nCleaned Clinician response (first entry):")
print(train_df['Cleaned_Clinician'].iloc[0])

    
```
